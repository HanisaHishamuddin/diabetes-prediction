```python
bins = [0, 18.5, 24.9, 29.9, float('inf')]
labels = ['underweight', 'normal_weight', 'overweight', 'obese']

#categorize BMI column
df['BMI_cat'] = pd.cut(df['BMI'], bins=bins, labels=labels, right=False)

df.drop('BMI', axis=1, inplace=True)
```

```python
VC_genhtlh = df.groupby('GenHlth').size()
print("Value counts in GenHealth:")
print(VC_genhtlh)
```

```python
#Count each unique value in MentalHealth column
VC_Mentalhealth = df.groupby('MentHlth').size()
print("Value counts in MentHealth:")
print(VC_Mentalhealth)
```

```python
#categorize GenHlth
df['MentHlth_cat'] = pd.cut(df['MentHlth'], bins=[0, 6, 12, 18, 24, 31], labels=['very_low', 'low', 'medium', 'high', 'very_high'], right=False)

# Drop the original 'MentalHealth' column
df.drop('MentHlth', axis=1, inplace=True)
```

```python
minmax_PhysHlth = df['PhysHlth'].describe()[['min', 'max']]

# Display the range of values for the 'PhysHlth' feature
print("Min_Max value range for PhysHlth:")
print(minmax_PhysHlth)
```

```python
#Count each unique value in PhysicalHealth column
VC_PhysicalHealth = df.groupby('PhysHlth').size()
print("Value counts in PhysicalHealth:")
print(VC_PhysicalHealth)
```

```python
# Assuming df1 is your DataFrame and 'PhysHlth' is the column you want to categorize
df['PhysHlth_cat'] = pd.cut(df['PhysHlth'], bins=[0, 6, 12, 18, 24, 31], labels=['Very_Low', 'Low', 'Medium', 'High', 'Very_High'], right=False)

# Drop the original 'PhysHlth' column
df.drop('PhysHlth', axis=1, inplace=True)


# Display the DataFrame with the new 'PhysHlth_CTG' column
df.head()
```

```python
#Count each unique value in Education column
VC_Education = df.groupby('Education').size()
print("Value counts in Education:")
print(VC_Education)
```

```python
#categorize Education
Education_cat = {
    1: 'Not Educated',
    2: 'Elementary School',
    3: 'High School',
    4: 'College',
    5: 'Bachelor Degree',
    6: 'Masters Degree'
}

df['Education_cat'] = df['Education'].map(Education_cat)

# Drop the original 'GenHlth' column
df.drop('Education', axis=1, inplace=True)

df.head()
```

```python
#Count each unique value in Age column
VC_Age = df.groupby('Age').size()
print("Value counts in Age:")
print(VC_Age)
```

```python
#Count each unique value in Income column
VC_Income = df.groupby('Income').size()
print("Value counts in Age:")
print(VC_Income)
```

```python
bins = [1, 2, 3, 4, 5, 6, 7, 8, 9]
labels = ['1', '2', '3', '4', '5', '6', '7', '8']

df['Income_cat'] = pd.cut(df['Income'], bins=bins, labels=labels, right=False)

df.drop('Income', axis=1, inplace=True)

df.head()
```

```python
df_encoded = pd.get_dummies(df)

df_encoded.head()
```

```python
df_ED_bool =df_encoded.astype(bool)

df_ED_bool.head()
```

```python
df_ED_bool.columns
```

```python
df_assco_rules =  df_ED_bool[['BMI_cat_underweight', 'BMI_cat_normal_weight',
       'BMI_cat_overweight', 'BMI_cat_obese', 'genhlth_cat_bad',
       'genhlth_cat_excellent', 'genhlth_cat_good', 'genhlth_cat_medium',
       'genhlth_cat_very_bad', 'MentHlth_cat_very_low', 'MentHlth_cat_low',
       'MentHlth_cat_medium', 'MentHlth_cat_high', 'MentHlth_cat_very_high',
       'PhysHlth_cat_Very_Low', 'PhysHlth_cat_Low', 'PhysHlth_cat_Medium',
       'PhysHlth_cat_High', 'PhysHlth_cat_Very_High',
       'Education_cat_Bachelor Degree', 'Education_cat_College',
       'Education_cat_Elementary School', 'Education_cat_High School',
       'Education_cat_Masters Degree', 'Education_cat_Not Educated',
       'Income_cat_1', 'Income_cat_2', 'Income_cat_3', 'Income_cat_4',
       'Income_cat_5', 'Income_cat_6', 'Income_cat_7', 'Income_cat_8']]
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
fp_apriori = apriori(df_ED_bool, min_support=0.3, use_colnames=True)

# Display the frequent itemsets and association rules
print("Frequent Itemsets:")
fp_apriori
```

```python
frequent_itemsets = apriori(df_ED_bool, min_support=0.1, use_colnames=True)

assoc_rules = association_rules(frequent_itemsets, metric='support', min_threshold=0.01)
display(assoc_rules)
```

```python
#Display support
from matplotlib import pyplot as plt
fp_apriori['support'].plot(kind='hist', bins=20, title='support')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
